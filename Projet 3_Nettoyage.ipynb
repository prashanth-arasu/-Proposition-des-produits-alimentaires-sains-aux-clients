Le Nutri-Score est calculé à partir de ces variables:

```python
# Outliers de la variable 'energy-kcal_100g'

# Remplacer les valeurs manquantes
data_app['energy_100g'] = data_app['energy_100g'].fillna(data_app['energy-kj_100g'])
data_app['energy-kcal_100g'] = data_app['energy-kcal_100g'].fillna(data_app['energy_100g'] / 4.184)

# Outliers 
data_app.loc[(data_app['energy-kcal_100g'] < 0), 'energy-kcal_100g'] = 0
data_app.loc[(data_app['energy-kcal_100g'] > 3700), 'energy-kcal_100g'] = 3700
```

```python
# Outliers de la variable 'fiber_100g'

# Outliers 
data_app.loc[(data_app['fiber_100g'] < 0), 'fiber_100g'] = 0
data_app.loc[(data_app['fiber_100g'] > 100), 'fiber_100g'] = 100
```

```python
# Outliers de la variable 'proteins_100g'

# Outliers 
data_app.loc[(data_app['proteins_100g'] < 0), 'proteins_100g'] = 0
data_app.loc[(data_app['proteins_100g'] > 100), 'proteins_100g'] = 100
```

```python
# Supprimer les variables redondantes

drop_cols_2 = ['allergens_en', 'additives_tags', 'additives_en', 'energy_100g', 'energy-kj_100g', 'sodium_100g', 'fruits-vegetables-nuts_100g', 'fruits-vegetables-nuts-dried_100g']
data_app = data_app.drop(columns =drop_cols_2, axis = 1)
```

Le Nutri-Score de chaque produit sera calculé en fonction des variables. L'attribution du score de chaque variable sera fait dans un nouveau dataframe: score_list

```python
# Définition du dataframe: score_list

column_names_2 = ['code','product_name', 'brands', 'allergens', 'additives', 'pnns_groups_2',
                  'energy', 'sugars', 'fat', 'salt', 'proteins', 'fruits', 'fiber', 'nutri_score']
score_list = pd.DataFrame(columns = column_names_2)
score_list['code'] = data_app['code']
score_list['product_name'] = data_app['product_name']
score_list['brands'] = data_app['brands']
score_list['allergens'] = data_app['allergens']
score_list['additives'] = data_app['additives']
score_list['pnns_groups_2'] = data_app['pnns_groups_2']
score_list
```

```python
# Attribution du Nutri-Grade pour chaque produit

solids = ['Appetizers', 'Artificially sweetened beverages', 'Biscuits and cakes', 'Bread', 'Breakfast cereals', 'Cereals',
          'Cheese', 'Chocolate products', 'Dairy desserts', 'Dressings and sauces', 'Dried fruits', 'Eggs', 'Fats',
          'Fish and seafood', 'Fruits', 'Ice cream', 'Legumes', 'Meat', 'Nuts', 'Offals', 'One-dish meals', 'Pastries',
          'Pizza pies and quiches', 'Potatoes', 'Processed meat', 'Salty and fatty products', 'Sandwiches', 'Sweets',
          'Vegetables']

liquids = ['Alcoholic beverages', 'Fruit juices', 'Fruit nectars', 'Milk and yogurt', 'Plant-based milk substitutes',
           'Soups', 'Sweetened beverages', 'Teas and herbal teas and coffees', 'Unsweetened beverages']

waters = ['Waters and flavored waters']
           
score_list.loc[(score_list['pnns_groups_2'].isin(waters)), 'nutri_grade'] = 'A'

score_list.loc[(score_list['pnns_groups_2'].isin(liquids) & score_list['nutri_score'].isin(range(-15, 2))), 'nutri_grade'] = 'B'
score_list.loc[(score_list['pnns_groups_2'].isin(liquids) & score_list['nutri_score'].isin(range(2, 6))), 'nutri_grade'] = 'C'
score_list.loc[(score_list['pnns_groups_2'].isin(liquids) & score_list['nutri_score'].isin(range(6, 10))), 'nutri_grade'] = 'D'
score_list.loc[(score_list['pnns_groups_2'].isin(liquids) & score_list['nutri_score'].isin(range(10, 41))), 'nutri_grade'] = 'E'

score_list.loc[(score_list['pnns_groups_2'].isin(solids) & score_list['nutri_score'].isin(range(-15, 0))), 'nutri_grade'] = 'B'
score_list.loc[(score_list['pnns_groups_2'].isin(solids) & score_list['nutri_score'].isin(range(0, 3))), 'nutri_grade'] = 'C'
score_list.loc[(score_list['pnns_groups_2'].isin(solids) & score_list['nutri_score'].isin(range(3, 11))), 'nutri_grade'] = 'D'
score_list.loc[(score_list['pnns_groups_2'].isin(solids) & score_list['nutri_score'].isin(range(11, 19))), 'nutri_grade'] = 'E'
score_list.loc[(score_list['pnns_groups_2'].isin(solids) & score_list['nutri_score'].isin(range(19, 41))), 'nutri_grade'] = 'E'

score_list_per_grade = score_list.sort_values(by = 'nutri_grade', ascending = True)

score_list
```
